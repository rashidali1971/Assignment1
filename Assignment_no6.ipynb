```python
# Q3. Assignment No 6

# Function
# A function is a block of code to carry out a specific task, will contain its own scope and is called by name. 
# All functions may contain zero(no) arguments or more than one arguments. 
# On exit, a function can or can not return one or more values.
# Method
# A method in python is somewhat similar to a function, except it is associated with object/classes. 
# Methods in python are very similar to functions except for two major differences.
# •	The method is implicitly used for an object for which it is called.
# •	The method is accessible to data that is contained within the class.
```

```python
# Q4. Assignment No 6

# Class
# A class is a user-defined blueprint or prototype from which objects are created. 
# Classes provide a means of bundling data and functionality together

# Objects
# An object(Instance)is an instatitiaon of a class.
# Object is simply a collection of data (variables) and methods (functions) that act on those data. 

# Attribute
# Attribute is object characteristic.
# A class attribute is a Python variable that belongs to a class rather than a particular object. 
# It is shared between all the objects of this class and it is defined outside the constructor function

# Behavior
# Behavior is object characteristic.
# Behavior binds the structure of objects to their attributes and relationships so that objects can meet their responsibilities
```

```python
# Q5. Assignment No 6

class Car(object):
    def __init__(self, company, model, color, passengers, speed):
        self.company    = company
        self.model      = model
        self.passengers = passengers
        self.color      = color
        self.speed      = speed
        
    def method1(self):
        print ("Model:",self.model)
        
    def method2(self):
        print ("Color:",self.color)

    def method3(self):
        self.speed = self.speed + 2
        print ("Speed limit:",self.speed,"\n")

    
# Instance 1. for BMW 
bmw = Car("BMW", 2019, "Red", 4, 5)
print ("Comapny: ",bmw.company)
bmw.method1()
bmw.method2()
print ("Passangers: ",bmw.passengers)
bmw.method3()

# Instance 2. for Toyota 
toyota = Car("Toyota", 2018, "Black", 4, 10)
print ("Comapny: ",toyota.company)
toyota.method1()
toyota.method2()
print ("Passangers: ",toyota.passengers)
toyota.method3()

# Instance 3. for Suzuki 
suzuki = Car("Suzuki", 2015, "White", 4, 15)
print ("Comapny: ",suzuki.company)
suzuki.method1()
suzuki.method2()
print ("Passangers: ",suzuki.passengers)
suzuki.method3()

# Instance 4. for Ford 
ford = Car("Ford", 2010, "Silver", 6, 20)
print ("Comapny: ",ford.company)
ford.method1()
ford.method2()
print ("Passangers: ",ford.passengers)
ford.method3()

# Instance 5. for Ferrari 
ferrari = Car("Ferrari", 2000, "Blue", 2,  25)
print ("Comapny: ",ferrari.company)
ferrari.method1()
ferrari.method2()
print ("Passangers: ",ferrari.passengers)
ferrari.method3()
```
